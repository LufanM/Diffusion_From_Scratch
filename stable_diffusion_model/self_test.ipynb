```python
import torch
import matplotlib.pyplot as plt

# 生成数据
x = torch.linspace(-5, 5, 1000)
relu = torch.relu(x)
silu = torch.nn.functional.silu(x)
gelu = torch.nn.functional.gelu(x)

# 绘图
plt.figure(figsize=(10, 6))
plt.plot(x.numpy(), relu.numpy(), label='ReLU', linewidth=3, color='#e74c3c', linestyle='-')
plt.plot(x.numpy(), silu.numpy(), label='SiLU (Swish)', linewidth=3, color='#3498db', linestyle='-')
plt.plot(x.numpy(), gelu.numpy(), label='GELU', linewidth=3, color='#2ecc71', linestyle='-')

# 美化
plt.axhline(0, color='black', linewidth=0.8, linestyle='--', alpha=0.3)
plt.axvline(0, color='black', linewidth=0.8, linestyle='--', alpha=0.3)
plt.grid(True, alpha=0.3, linestyle=':', linewidth=0.5)
plt.legend(fontsize=14, loc='upper left', framealpha=0.9)
plt.xlabel('Input (x)', fontsize=14, fontweight='bold')
plt.ylabel('Output f(x)', fontsize=14, fontweight='bold')
plt.title('Activation Functions: ReLU vs SiLU vs GELU', fontsize=16, fontweight='bold', pad=20)
plt.xlim(-5, 5)
plt.ylim(-1, 5)

# 标注关键区域
plt.fill_between(x.numpy(), -1, 0, where=(x.numpy() < 0), alpha=0.1, color='red', label='Negative Region')
plt.text(-3, 3.5, 'ReLU: Hard cutoff', fontsize=10, color='#e74c3c', fontweight='bold')
plt.text(-3, 3.0, 'SiLU/GELU: Smooth', fontsize=10, color='#3498db', fontweight='bold')

plt.tight_layout()
# plt.savefig('activation_simple.png', dpi=300, bbox_inches='tight', facecolor='white')
plt.show()

print("✅ 简化版图片已保存")
```
